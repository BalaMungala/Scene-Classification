```python

#  Building the CNN

# Importing the Keras libraries and packages
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense




classifier = Sequential()

# Step 1 - Convolution
classifier.add(Conv2D(32, (3, 3), input_shape = (64, 64, 3), activation = 'relu'))

# Step 2 - Pooling
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Adding a second convolutional layer and max pooling
classifier.add(Conv2D(32, (3, 3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))

# Step 3 - Flattening
classifier.add(Flatten())

# Step 4 - Full connection
classifier.add(Dense(units = 128, activation = 'relu'))
classifier.add(Dense(units = 6, activation = 'softmax'))

classifier.load_weights("E://Analytics Vidhya//Intel Scene//weights//intelimg_classify_weights.03-0.08.hdf5")
```

```python
# Making new predictions

import numpy as np
import pandas as pd
from keras.preprocessing import image
from pandas import DataFrame

dd=pd.read_csv("E://Analytics Vidhya//Intel Scene//sample_submission_CH2mq5Z.csv")

test=[]
a=0
b=1
c=2
d=3
e=4
f=5
k=6

for i in dd['image_name'].values:
        test_image = image.load_img('E://Analytics Vidhya//Intel Scene//train-scene classification//test_images//{}'.format(i), target_size = (64, 64))
        test_image = image.img_to_array(test_image)
        test_image = np.expand_dims(test_image, axis = 0)
        result = classifier.predict(test_image)
        dataa = {'label':[0,1,2,3,4,5],'output':result.ravel()}
        rr=DataFrame(dataa)
        if(rr.iloc[rr['output'].values==1]['label'].ravel()==0):
                    test.append(a)  
                    print("building")
        elif(rr.iloc[rr['output'].values==1]['label'].ravel()==1):
                    test.append(b)  
                    print("forest")
        elif(rr.iloc[rr['output'].values==1]['label'].ravel()==2): 
                    test.append(c)
                    print("glacier")
        elif(rr.iloc[rr['output'].values==1]['label'].ravel()==3):
                    test.append(d)  
                    print("mountain")
        elif(rr.iloc[rr['output'].values==1]['label'].ravel()==4):
                    test.append(e)  
                    print("sea")
        elif(rr.iloc[rr['output'].values==1]['label'].ravel()==5):
                    test.append(f)  
                    print("street")
        else:
                    dffrnt=rr.iloc[rr['output'].values==max(rr['output'])]['label'].ravel()
                    test.append(dffrnt[0])
                    print(dffrnt.ravel())
        
```
